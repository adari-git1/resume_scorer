```python
from pyspark.sql import SparkSession
```

```python
s1=SparkSession.builder.appName("Resume").getOrCreate()
```

```python
def resume_optimization(resume_as_text_file,key_words):
  count=0
  keys_present=[]
  keys_absent=[]
  rdd=s1.sparkContext.textFile(resume_as_text_file)
  key_words=[word.lower() for word in key_words]
  for key in key_words:
    filtered_rdd = rdd.filter(lambda line: key in line.lower()).count()
    if filtered_rdd>0:
      count+=1
      keys_present.append(key)
    else:
      keys_absent.append(key)
  print("Resume Match Score is",count/len(key_words)*100)
  print("Missing Skills are",keys_absent)
```

```python
resume="/content/file_name.txt"
key_words=["Python",
"SQL",
"data science",
"machine learning",
"advanced analytics",
"risk management",
"statistics",
"critical thinking",
"communication skills",
"problem-solving",
"presentation skills",
"verbal communication",
"written and verbal communication skills"]
resume_optimization(resume,key_words)
```
